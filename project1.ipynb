Relationship Between Visibility and Precipitation: The scatter plot shows a weak negative correlation between visibility and precipitation. This means that there is a slight tendency for lower visibility to occur during times with higher precipitation, and vice versa.

This relationship makes sense because rain reduces visibility due to water droplets in the air.

```python
import numpy as np

np.random.seed(0)
date_rng = pd.date_range(start='1/1/2023', end='31/12/2023', freq='D')
humidity = np.random.uniform(low=20, high=100, size=(len(date_rng))) 
visibility = np.random.uniform(low=1, high=10, size=(len(date_rng)))  


sc2 = pd.DataFrame(date_rng, columns=['datetime'])
sc2['humidity'] = humidity
sc2['visibility'] = visibility
sc2['datetime'] = pd.to_datetime(sc2['datetime'])
sc2.set_index('datetime', inplace=True)

plt.figure(figsize=(15, 5))
plt.scatter(sc2['humidity'], sc2['visibility'], alpha=0.6)
plt.xlabel('Humidity (%)')
plt.ylabel('Visibility (km)')
plt.title('Scatter Plot between Humidity and Visibility')
plt.show()
```

```python
columns_to_analyze = ['tempmax', 'tempmin', 'windspeed', 'visibility']
statistics = {}
for column in columns_to_analyze:
    column_data = df[column]
    statistics[column] = {
        'median': column_data.median(),
        'mean': column_data.mean(),
        'max': column_data.max(),
        'min': column_data.min()
    }
for column, stats in statistics.items():
    print(f"Statistics for {column}:")
    print(f"  Median: {stats['median']}")
    print(f"  Mean: {stats['mean']}")
    print(f"  Max: {stats['max']}")
    print(f"  Min: {stats['min']}")
    print()
```
